```python
# 3) Elabore um código que peça dois números e imprima a soma.
print("Digite dois números")
numero1 = int(input())
numero2 = int(input())
print("A soma dos números é", numero1 + numero2)
```

```python
# 4) Elabore um código que peça as 4 notas bimestrais e mostre a média.
print("Digite as 4 notas bimestrais")
nota1 = float(input())
nota2 = float(input())
nota3 = float(input())
nota4 = float(input())
print("A média das notas é", (nota1 + nota2 + nota3 + nota4) / 4)
```

```python
#outro codigo da tarefa numero 4) Elabore um código que peça as 4 notas bimestrais e mostre a média.
notas = []

for i in range(1, 5):
    while True:
        try:
            nota = float(input(f"Digite a {i}ª nota (0-10): "))

            if 0.00 <= nota <= 10.00:
                notas.append(nota)
                break
            else:
                print("Valor inválido! Por favor, digite uma nota entre 0.00 e 10.")
        except ValueError:
            print("Entrada inválida! Digite um número.")

soma = sum(notas)
media = soma / 4

print("-" * 10)
print(f"A soma das notas é: {soma}")
print(f"A média final é: {media:.2f}")
if media >= 5:
    print("Aprovado!")
else:
    print("Reprovado!")
```

```python
# 5) Elabore um código que converta metros para centímetros.
metros = float(input("Digite o valor em metros: "))
centimetros = metros * 100
print(f"{metros} metros equivalem a {centimetros:.0f} centímetros.")
```

```python
# 6) Elabore um código que peça o raio de um círculo, calcule e mostre sua área.
raio = float(input("Digite o raio do círculo: "))
area = 3.14 * (raio ** 2)
print(f"A área do círculo é: {area:.2f}")
```

```python
# 7) Elabore um código que calcule a área de um quadrado, em seguida mostre o dobro desta área para o usuário.
lado = float(input("Digite o valor do lado do quadrado: "))
area = lado
dobro_area = area * 2
print(f'A area do quadrado é:{area}')
print(f'O dobro da area é:{dobro_area}')
```

```python
# 8) Elabore um código que pergunte quanto você ganha por hora e o número de horas trabalhadas no mês.
#Calcule e mostre o total do seu salário no referido mês em uma mensagem amigável.
valor_hora = float(input("Digite o valor que você ganha por hora: "))
horas_trabalhadas = float(input("Digite o número de horas trabalhadas no mês: "))
salario = valor_hora * horas_trabalhadas
print(f"Seu salário no referido mês é: R${salario:.2f}")
```

```python
# 9) Elabore um código que peça a temperatura em graus Fahrenheit, transforme e mostre a temperatura em graus Celsius.
# ** Onde C = 5 * ((F-32) / 9)**.
fahrenheit = float(input("Digite a temperatura em graus Fahrenheit: "))
celsius = 5 * ((fahrenheit - 32) / 9)
print(f"{fahrenheit} graus Fahrenheit equivalem a {celsius:.2f} graus Celsius.")
```

```python
# 10) Elabore um código que peça a temperatura em graus Celsius, transforme e mostre em graus Fahrenheit.
celsius = float(input("Digite a temperatura em graus Celsius: "))
fahrenheit = (celsius * 9/5) + 32
print(f"{celsius} graus Celsius equivalem a {fahrenheit:.2f} graus Fahrenheit.")
```

```python
# 11) Elabore um código que peça 2 números inteiros e um número real. Calcule e mostre:
#o produto do dobro do primeiro com metade do segundo .
#a soma do triplo do primeiro com o terceiro.
#o terceiro elevado ao cubo.
print(f'Digite dois números inteiros')
numero1 = int(input())
numero2 = int(input())
print(f'Digite um número real')
numero3 = float(input())
produto = (numero1 * 2) * (numero2 / 2)
soma = (numero1 * 3) + numero3
elevado = (numero3 ** 3)
print(  f'O produto do dobro do primeiro com metade do segundo é: {produto}')
print(f'A soma do triplo do primeiro com o terceiro é: {soma}')
print(f'O terceiro elevado ao cubo é: {elevado}')
```

```python
# 12) Elabore um código que receba os dados de entrada a altura de uma pessoa, e informe qual seria o peso ideal, usando a seguinte fórmula:
# (72.7*altura) - 58
print(f'Digite a sua altura.')
altura = float(input())
altura = (72.7 * altura) - 58
print(f'O seu peso ideal é: {altura:.2f}')
```
