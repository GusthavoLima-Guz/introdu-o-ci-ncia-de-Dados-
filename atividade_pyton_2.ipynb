```python
# Elabore uma função chamada calcula_desconto que receba dois parâmetros: o valor de um produto e a porcentagem de desconto.
# A função deve retornar o valor final do produto com o desconto aplicado. Teste a função e imprima o resultado com input.

def calcula_desconto(valor_produto, porcentagem_desconto):
    valor_desconto = valor_produto * (porcentagem_desconto / 100)
    valor_final = valor_produto - valor_desconto
    return valor_final

valor_original_input = float(input('Digite o valor original do produto: '))
porcentagem_desconto_input = float(input('Digite a porcentagem de desconto: '))

valor_com_desconto_input = calcula_desconto(valor_original_input, porcentagem_desconto_input)
print(f"O valor original do produto é R${valor_original_input:.2f}.")
print(f"Com um desconto de {porcentagem_desconto_input:.0f}%, o valor final é R${valor_com_desconto_input:.2f}.")
```

```python
# Elabore uma função chamada classifica_nota que receba uma nota de 0 a 10. A função deve retornar 'Aprovado' se a nota for maior ou igual a 7,
#'Recuperação' se estiver entre 5 e 6.9, e 'Reprovado' se for menor que 5. Chame a função passando uma nota e imprima o retorno.

def classifica_nota(nota):
    if nota >= 7:
        return "Aprovado"
    elif 5 <= nota < 7:
        return "Recuperação"
    else:
        return "Reprovado"
nota_usuario = float(input('Digite a nota (de 0 a 10): '))
resultado_usuario = classifica_nota(nota_usuario)
print(f"A nota {nota_usuario} é classificada como: {resultado_usuario}")
```

```python
#Elabore um código que receba uma lista de números inteiros mistos (positivos, negativos e zeros).
#Utilize um laço for com estruturas condicionais if/elif/else para contabilizar quantos números são positivos,
# quantos são negativos e quantos são exatamente zero. Ao final, imprima esse resumo.

entrada_numeros = input('Digite uma lista de números inteiros separados por vírgula (ex: 1,-2,0,5): ')

try:
    lista_numeros_mistos = [int(num.strip()) for num in entrada_numeros.split(',')]
except ValueError:
    print("Entrada inválida. Por favor, digite apenas números inteiros separados por vírgula.")
    lista_numeros_mistos = []

positivos = 0
negativos = 0
zeros = 0

for numero in lista_numeros_mistos:
    if numero > 0:
        positivos += 1
    elif numero < 0:
        negativos += 1
    else:
        zeros += 1

print(f"Na lista, temos:")
print(f"- {positivos} números positivos")
print(f"- {negativos} números negativos")
print(f"- {zeros} números zero")
```
